```python
#IMPORTANDO BIBLIOTECAS
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_videos = pd.read_excel('videosYT.xlsx','videos')
display (df_videos)
```

```python
#CRIANDO UM GRAFICO DE DISPERSÃO
```

```python
fig = sns.scatterplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue='Responsável', palette = ['red', 'black'])

plt.show(fig)
```

```python
#OUTRA FORMA DE FAZER
```

```python
grafico_real = sns.relplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue='Categoria', col='Responsável')
grafico_real.set_titles('Responsável:{col_name}')
plt.show(grafico_real)
```

```python
#GRAFICO DE LINHA
```

```python
df_inscritos = pd.read_excel('videosYT.xlsx','Inscritos')
display (df_inscritos)

graf_linha = sns.lineplot(data=df_inscritos, x='Mês/Ano', y = 'Inscritos', color = 'red')
plt.show(graf_linha)
```

```python
#CRIANDO HISTOGRAMAS
```

```python
fig = sns.displot(data = df_videos, x='Nº de Views', hue='Responsável', col='Responsável')
plt.show(fig)
```

```python
#REGRAÇÃO LINEAR
```

```python
graf_regr = sns.regplot(data=df_videos, x = 'Nº de Views' , y = 'Nº de Likes' )
plt.show(graf_regr)
```

```python
draf_regr = sns.lmplot(data=df_videos, x = 'Nº de Views' , y = 'Nº de Likes', hue = 'Responsável', markers = ['o','x'])
plt.show(graf_regr)
```
